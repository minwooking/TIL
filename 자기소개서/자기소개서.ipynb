#### 포기하지않고 도전한다


```python
- 참가인원 제한이 1 ~ 3명이였고 저도 팀원을 모아서 해커톤에 지원을 하려고 제안서를 적었지만 당시 당일날 (취업,학업)등의 이유로 팀은 와해되어 해커톤을 포기해야 되는 상황이 오게되었습니다. 
- 당일날 안하겠다는 팀원들에 의해서 혼자서 출전을 하거나 포기해야 되는 상황이였고 저는 그 때 출전을 도전했습니다. 아마 그때는 서류 통과만 되도 괜찮다 라는 생각으로 시도 했던 것 같습니다.
```

#### 선택 과 집중 적절한 계획 설정

```python
- 이후 제안서를 마무리 하여 제출한 이후 본선을 통과 하였고 게더타운을 통해서 한자리에 모여있을때는 유일한 개인팀이였습니다.
- 많이 착잡했습니다. 저를 제외한 모든 조가 대다수가 3명 이였고 개인팀은 혼자였기 때문입니다.
- 하지만 해보면 배울게 많을 것이다. 혼자서 해보는것도 경험이고 이걸 전화위복으로 삼자 라는 생각을 하고 다잡고 도전했습니다.
- 이겨낸 방법 

1. 먼저 본래 아이디어와 실현 계획을 구체적으로 설정한다.
    ㅇ 시간을 신경쓰지 않고 원래 생각했던 아이디어를 구체적으로 정리하였습니다.
2. 객관적으로 판단하여 포텐셜을 설정한다.
    ㅇ 아이디어 구현에서 목표를 심사위원들에게 아이디어 실현의 가능성을 보여주자로 재설정하였습니다.
    ㅇ 위에서 정리한 계획을 주어진 시간(약 2주) 인력(개인 1명)이 최대로 큰 성과를 낼 수 있는 계획을 재설정 하였습니다.
    - 데이터 분석 및 가공 (2일) = > 딥러닝(image) 및 머신러닝(json파일) (3일) = > 앱제작 및 앱에 Tfmodel 이식(3일) => 앱 기능추가 (2일) = > NFT 제작 및 자동화 알고리즘화(0.5일) => PPT 제작(1일)

위 계획처럼 계획을 나눠서 시간이 초과될 경우에는 다음 목표를 처리했습니다.(폭포수)
```

## 3. 회사에 기여할 수 있거나 기여하고 싶은 부분


#### 제주도 토박이

```python
ㅇ 저는 제주도 토박이입니다. 해병대 정보통신장교로 김포 전방에서 2년 근무한걸 제외하면 제주도에 항상있었습니다.
이를 통하여 실차실험시 장소 및 협조 선정을 잘 할 수 있습니다.

ㅇ 팀원들에게 좀 더 유익한 제주도 생활 라이프를 제공 할 수 있습니다.
제주도 도민들만 알 수 있는 여러 장소를 팀원들과 공유하여 팀의 사기를 올리겠습니다

ㅇ 뛰어난 커뮤니케이션 능력
통신소대장 근무시절 소대원들과 간부들의 관리하기 위하여 커뮤니케이션 능력을 스스로 개발시켰습니다.
대학시절 또래상담사 과정과 또래상담사리더 과정을 통해서 타인에게 공감하고 경청하는 법등을 전문적으로 익혔습니다.

ㅇ 업무 관리 / 일정 관리 
지인의 부탁으로 잠시동안 KTCS(디지털배움터)에서 근무하면서 교육 일정과 교육업무를 관리 하는 업무를 하였습니다. 이를 통하여 실질적인 일정 계획을 할 수 있다고 생각합니다.
실제와 업무 및 일정을 맞게 계획 할 수 있습니다. 전방에서 작전지역의 통신지원을 하면서 텍스트로는 다 계획 할 수 없는 계획들이 많다는걸 알게 되었습니다. 완벽 할 수는 없지만
최대한 실제가 업무 및 일정에 차질이 생기지 않게 계획 할 자신이 있습니다.

ㅇ 포기하지 않는 끊기
큰 성과는 가장 힘들 때 나온다고 생각합니다. 해커톤을 통해서 확신을 얻게 되었습니다. 이를 회사에 적용 할 수 있다고 생각합니다.

ㅇ 일단 해보는 패기
부족하면 수정하면 된다고 생각합니다. 많은 아이디어와 계획을 회사를 위해서 만들고 의견 내는걸 두려워 하지 않겠습니다.

ㅇ 수많은 경험
저는 나이에 비해서 많은 경험을 했다고 생각합니다. 이 경험을 통해서 수많은 아이디어가 나올 수 있다고 생각합니다. 

참고 
포트폴리오 : 첨부물(자기소개서.pdf)
```
